```python
import os
import numpy as np
import pandas as pd
import cv2
import colorsys
```

```python
# given a folder full of color blocks
# return the average r, g, b, and l values
# r, g, b range from 0 - 255
# l ranges from 0 - 1

def get_pixels(folder):
    
    pixels = []
    for image_name in os.listdir(folder):
        
        if ("color" not in image_name):
            continue
            
        image = cv2.imread(folder + "/" + image_name)
        pixel = image[0][0]
        pixels.append(pixel)
    
    return pixels
    
```

```python
# loop through the whole folder and create a dataframe

mali = "all_pictures/compressed/mali"
mandi = "all_pictures/compressed/mandi"
out = ["volume,issue,year,r,g,b,l,file_name"]

for vol_issue in os.listdir(mali):
    print(vol_issue)
    malis_pixels = get_pixels(mali + "/" + vol_issue + "/0valid")
    mandis_pixels = get_pixels(mandi + "/" + vol_issue + "/0valid")
    pixels = np.array(malis_pixels + mandis_pixels)
    [b, g, r] = np.average(pixels, axis=0)
    (h, l, s) = colorsys.rgb_to_hls(r / 255, b / 255, g / 255)
    vol, issue, year, face = vol_issue.split("_")
    row = ",".join([vol, issue, year, str(int(r)), str(int(g)), str(int(b)), str(l), vol_issue])
    out.append(row)
    
```

```python
# combine the two tables into one

with open("all.csv") as f:
    lines = f.read().split("\n")
    
out_all = []
for (new, old) in zip(out, lines):
    model_name = old.split(",")[7]
    new_line = new + "," + model_name
    out_all.append(new_line)
    
with open("all1.csv", "w") as f:
    data = "\n".join(out_all)
    f.write(data)
```
